```python
import numpy as np
import matplotlib.pyplot as plt
import torch
from koopman.simulation.systems import NonlinearAttractor2D
from koopman.simulation.simulator import simulate
```

```python
attractor = NonlinearAttractor2D(NonlinearAttractor2D.Params(mu=2.0, lam=0.5))
ts, xhist, uhist = simulate(
    sys=attractor,
    tf=8.0,
    dt=0.05,
    x0=np.array([10.0, 10.0]),
    u=lambda t, x: np.zeros(1)
)
```

```python
plt.plot(xhist[:, 0], xhist[:, 1], label='Path')
plt.scatter(xhist[0, 0], xhist[0, 1], label='Start', color='red')
plt.scatter(xhist[-1, 0], xhist[-1, 1], label='End', color='green')

plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
plt.plot(ts, xhist[:, 0], label='x')
plt.plot(ts, xhist[:, 1], label='y')

plt.xlabel('t')
plt.ylabel('states')
plt.legend()
plt.show()
```
